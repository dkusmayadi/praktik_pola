Membaca file dan menyajikan dalam bentuk grafik
Untuk dapat menyajikan dalam bentuk grafik terlebih dahulu melakukan instalasi paket matplotlib

```python
import numpy as sp
traffic = sp.genfromtxt("web_traffic.tsv", delimiter='\t')
print(traffic[:10])
print(traffic.shape)

x = traffic[:, 0]
y = traffic[:, 1]

x = x[~sp.isnan(y)]
y = y[~sp.isnan(y)]

import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.title("Web Traffic Last Month")
plt.xlabel("Time")
plt.ylabel("Hits/hour")

plt.xticks([w*7*24 for w in range(10)], ['week %i' %w for w in range(10)])
plt.autoscale(tigh=true)
plt.grid()
```
